## Load Team Dataset

## Load Test Dataset

```python
#Loading data
test_data_frame = pd.read_csv('./testDataNoLabels.csv', header=None, index_col = False, names=dataTitles)

#Remove first row with column indexes
test_data_frame = test_data_frame.drop(0, axis=0)

#Show data_frame
test_data_frame
```

```python
#Functions for KNN

def predict(X_train, Y_train, X_test, k):
    #Create list for distances and k_closest
    distances = []
    k_closest = []

    for i in range(len(X_train)):
        #First we compute the euclidean distance
        distance = np.sqrt(np.sum(np.square(X_test - X_train[i, :])))
        #Add it to list of distances
        distances.append([distance, i])
    
    #Sort the list
    distances = sorted(distances)

    #Make a list of the k neighbors' targets
    for i in range(k):
        index = distances[i][1]
        k_closest.append(Y_train[index])
    
    #Return most common out of k_closest
    prediction, _ = stats.mode(k_closest)
    return prediction


def kNearestNeighbor(X_train, Y_train, X_test, k):
    #Stores predictions 
    predictions = []

    #Loop over all observations
    for i in range(len(X_test)):
        ith_prediction = predict(X_train, Y_train, X_test[i, :], k)
        predictions.append(ith_prediction)
        
    #Remove extra dimension
    predictions = np.squeeze(predictions, axis=2)
    
    return predictions
```

#### The code below is what you will change to modify your algorithm:

```python
#Get the first 150 columns from the team's dataframe, this is the length of our feature vector
X_train = np.array(team_data_frame[team_data_frame.columns[:150]])

Y_train = np.array(team_data_frame[['Encoding']])


#Get the first 150 columns from the test dataframe, this is the length of our feature vector
X_test = np.array(test_data_frame[test_data_frame.columns[:150]])


#____________________________Run KNN_________________________________
#Compute KNN and obtain predictions
k = 1
predictions = kNearestNeighbor(X_train, Y_train, X_test, k)

#Show our predictions and labels
print('Prediction encodings:\n', predictions)
print('\n')

#Get the name of the label the algorithm produced
predictionLabels = []
for p in (predictions):
    label = np.array(team_data_frame.loc[team_data_frame['Encoding'] == p])
    label = label[1,151]
    predictionLabels.append(label)
    

for i in range (0,len(predictionLabels)):
    print('Prediction ' + str(i+1) +  ': ', predictionLabels[i])
    
```

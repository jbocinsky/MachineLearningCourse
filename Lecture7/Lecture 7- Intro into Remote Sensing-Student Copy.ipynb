# What is Remote Sensing?

According to the United States Geological Survey (USGS):

Remote sensing is the process of detecting and monitoring the physical characteristics of an area by measuring its reflected
and emitted radiation at a distance from the targeted area. Special cameras collect remotely sensed images of the Earth, 
which help researchers "sense" things about the Earth$^1$

# Examples of Remote Sensing

-Cameras on satellites and airplanes (Hyperspectral data)

-Sonar systems on ships (SONAR data)

# Introduction to Hyperspectral Data 

What is Hyperspectral Data?

According to Dr. Shippert$^3$:
The hyper in hyperspectral means over as in too many and refers to the large
number of measured wavelength bands.  Hyperspectral images are spectrally
overdetermined, which means that they provide ample spectral information to identify
and distinguish spectrally unique materials.

# Hyperspectral Data and Machine Learning



```python
# Display bands of different datasets
import matplotlib.pyplot as plt
import numpy as np

#Load datasets
SanBar = np.load('SanBarHyperIm.npy')#178 bands
Pavia = np.load('PaviaHyperIm.npy')#103 bands

#Select which band you want to display
SanBar_band = 100 #Needs to be between 0 and 177
Pavia_band = 102  #Needs to be between 0 and 102

#Plot images
plt.figure(figsize=(20,7))
plt.imshow(SanBar[:,:,SanBar_band])
plt.suptitle('Santa Barbara Band {}'.format(SanBar_band),fontsize=50)
plt.xlabel('X',fontsize=30)
plt.ylabel('Y',fontsize=30)

plt.figure(figsize=(20,10))
plt.imshow(Pavia[:,:,Pavia_band].T)
plt.suptitle('Pavia Band {}'.format(Pavia_band),fontsize=50)
plt.xlabel('X',fontsize=30)
plt.ylabel('Y',fontsize=30)
```

```python
#Plot Spectra
#Reshape HSI image to be Pixels by Bands
SanBar_Spectra = np.reshape(SanBar,(SanBar.shape[0]*SanBar.shape[1],SanBar.shape[2]))
Pavia_Spectra =  np.reshape(Pavia,(Pavia.shape[0]*Pavia.shape[1],Pavia.shape[2]))

#Select which pixel you want to view
SanBar_Pixel = 1000 #Needs to be between 0 and 32499
Pavia_Pixel = 990 #Needs to be between 0 and 207399

#Plot pixels
plt.figure(figsize=(25,7))
plt.figure(1)
plt.subplot(221)
plt.plot(SanBar_Spectra[SanBar_Pixel,:])
plt.xlabel('Band',fontsize=15)
plt.ylabel('Reflectance',fontsize=15)
plt.title('Santa Barbara Spectra of Pixel {}'.format(SanBar_Pixel),fontsize=20)

plt.subplot(222)
plt.plot(Pavia_Spectra[Pavia_Pixel,:])
plt.xlabel('Band',fontsize=15)
plt.ylabel('Reflectance',fontsize=15)
plt.title('Pavia Spectra of Pixel {}'.format(Pavia_Pixel),fontsize=20)
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=.2)
plt.show()
```
